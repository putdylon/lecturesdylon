```python
cols_to_drop = ["Rating"]
df = df.drop(cols_to_drop, axis=1)
```

```python
import matplotlib.pyplot as plt

# Data
sales_data = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
sales_totals = sales_data.sum()

# Membuat diagram batang
plt.figure(figsize=(10, 6))  # Ukuran diagram
bar_chart = sales_totals.plot(kind='bar', alpha=0.7)
plt.xlabel('Wilayah Penjualan')
plt.ylabel('Total Penjualan')
plt.title('Total Penjualan Menurut Wilayah')

# Menambahkan label angka jumlah di atas setiap batang
for i, value in enumerate(sales_totals):
    bar_chart.text(i, value, str(value), ha='center', va='bottom')

# Menampilkan diagram batang
plt.show()

# Menemukan wilayah dengan penjualan tertinggi
max_sales_region = sales_totals.idxmax()
print("Wilayah dengan penjualan tertinggi:", max_sales_region)
```

```python
import matplotlib.pyplot as plt

# Data
genre_sales_na = df.groupby('Genre')['NA_Sales'].sum()
genre_sales_eu = df.groupby('Genre')['EU_Sales'].sum()
genre_sales_jp = df.groupby('Genre')['JP_Sales'].sum()
genre_sales_other = df.groupby('Genre')['Other_Sales'].sum()
genre_sales_global = df.groupby('Genre')['Global_Sales'].sum()

# Membuat scatter plot untuk wilayah penjualan
plt.figure(figsize=(12, 6))
plt.scatter(genre_sales_na.index, genre_sales_na.values, alpha=0.5, label='NA Sales')
plt.scatter(genre_sales_eu.index, genre_sales_eu.values, alpha=0.5, label='EU Sales')
plt.scatter(genre_sales_jp.index, genre_sales_jp.values, alpha=0.5, label='JP Sales')
plt.scatter(genre_sales_other.index, genre_sales_other.values, alpha=0.5, label='Other Sales')
plt.scatter(genre_sales_global.index, genre_sales_global.values, alpha=0.5, label='Global Sales')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.title('Scatter Plot: Genre vs. Total Sales in Different Regions')
plt.xticks(rotation=90)
plt.legend()  #kasihn info

# Menampilkan scatter plot
plt.show()

```

```python
import matplotlib.pyplot as plt
]import seaborn as sns

# Data
sales_columns = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
genre_counts = df.groupby(sales_columns).count()['Genre']

# Menggabungkan genre_counts dengan data penjualan
correlation_data = df[sales_columns].join(genre_counts, on=sales_columns)

# Membuat matriks korelasi
correlation_matrix = correlation_data.corr()

# Membuat heatmap matriks korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=.5)
plt.title('Correlation Heatmap: Genre Count vs. Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
sales_columns = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
correlation_data = df.groupby('Year_of_Release')[sales_columns].sum()

# Membuat matriks korelasi
correlation_matrix = correlation_data.corr()

# Membuat heatmap matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=.5)
plt.title('Correlation Heatmap: Year_of_Release vs. Sales')
plt.show()
```

Jika nilai korelasi adalah 1, ini menunjukkan hubungan positif sempurna antara dua variabel. Artinya, jika salah satu variabel naik, yang lain juga naik dalam proporsi yang sama.

Jika nilai korelasi adalah -1, ini menunjukkan hubungan negatif sempurna antara dua variabel. Artinya, jika salah satu variabel naik, yang lain turun dalam proporsi yang sama.

Jika nilai korelasi mendekati 0 (positif atau negatif), ini menunjukkan bahwa tidak ada hubungan linear yang kuat antara dua variabel. Mereka cenderung tidak berkorelasi.

Jika nilai korelasi mendekati 1 (positif) atau -1 (negatif) tetapi tidak mencapai angka tersebut, ini menunjukkan hubungan positif atau negatif yang kuat, tetapi tidak sempurna. Ini berarti ada hubungan antara dua variabel, tetapi fluktuasi masih mungkin.

```python
# Data
sales_columns = ['NA_Sales', 'EU_Sales', 'JP_Sales','Other_Sales', 'Global_Sales']

top_combinations = {}
for region in sales_columns:
    # Menghitung total penjualan per genre, tahun rilis, dan wilayah
    total_sales_by_genre_year = df.groupby(['Genre', 'Year_of_Release'])[region].sum()

    # Menemukan top 3 kombinasi genre, tahun rilis, dan wilayah dengan total penjualan tertinggi
    top_combinations[region] = total_sales_by_genre_year.reset_index().sort_values(by=region, ascending=False).head(3)

# Menampilkan hasil per wilayah
for region, combinations in top_combinations.items():
    print(f"Top 3 Genre in {region}:")
    for index, row in combinations.iterrows():
        genre, year, total_sales = row['Genre'], row['Year_of_Release'], row[region]
        print(f"Genre: {genre}, Year of Release: {year}, Total {region} Sales: {total_sales}")
    print()
```
